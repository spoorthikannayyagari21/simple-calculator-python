```python
n1=float(input("enter value of n1"))
operator=input("enter the operator")
n2=float(input("enter value of n2"))
if operator=='+':
    print("n1+n2=",n1+n2)
elif operator=='-':
    print("n1-n2=",n1-n2)
elif operator=='x':
    print("n1xn2=",n1*n2)
elif operator=='/':
    print("n1/n2=",n1/n2)
elif operator=='//':
    print("n1//n2=",n1//n2)
else:
    print("error")

    
```
